## Câu 1: Vẽ bàn cờ 8x8 (0.5đ)
### Yêu cầu : Trên bàn cờ chỉ có ô đen (B) và ô trắng (W) đan xen nhau, với dòng đầu tiên trên bàn cờ, ô bắt đầu sẽ là ô trắng, ô tiếp theo là ô đen. Dòng thứ 2 ô đầu tiên sẽ là ô đen, ô tiếp theo là ô trắng. Và theo nguyên tắc đó cho cả bàn cờ.
### Hãy viết chương trình in bàn cờ 8x8.
image.png

```r
CreateMatrix <- function(size){
    mat = matrix(, nrow = size, ncol = size)

    for(row in 1:nrow(mat))
    {
        for(col in 1:ncol(mat))
        {
            if((row+col) %% 2 == 0)
                mat[row, col] = '| W |'
            else
                mat[row, col] = '| B |'
            
        }
    }
    return (mat)
}
```

```r
mat = CreateMatrix(8)
for(row in 1:nrow(mat))
{
    print(mat[row,])
    print('-------------------------------------------------------------')
    
}
```

## Câu 2: Vector (0.5đ)
### Yêu cầu : Viết chương trình xử lý vector như sau
1. Chương trình sẽ tạo ra 2 vector : vec1 và vec2 như sau :
    -  Người dùng nhập số lượng phần tử cho vec1, vec2 lần lượt là n, m
    -  Hệ thống tạo ra n phần tử số nguyên ngẫu nhiên cho vec1, m phần tử số nguyên ngẫu nhiên cho vec2
2. Cho biết những phần tử chỉ có trong vec1 mà không có trong vec2
3. Cho biết những phần tử xuất hiện trong cả vec1 và vec2
4. Tạo 1 vector mới là vec12 chứa tất cả các phần tử của vec1 và vec2. Thống kê số lần xuất hiện của từng phần tử. Phần tử nào xuất hiện nhiều lần nhất ?
5. Tạo 1 vector mới là vec_unique chỉ chứa những phần tử duy nhất từ vec12
### Ví dụ :
- Người dùng nhập n = 3
- Người dùng nhập m= 2
- Hệ thống tạo ra: vec1 = c(23, 18, 7), vec2 = c(15, 18)
- Phần tử xuất hiện trong cả vec1 và vec2 là: 18
- vec12 = c(23, 18, 7, 15, 18)
#### Thống kê:
- 23: xuất hiện 1 lần
- 18: xuất hiện 2 lần
- 7: xuất hiện 1 lần
- 15: xuất hiện 1 lần
- Phần tử xuất hiện nhiều nhất là 18 với 2 lần
#### vec_unique = c(23, 18, 7, 15)

```r
# n = 6
# m = 8

n = as.integer(readline('nhập số lượng phần tử vec1: '))
m = as.integer(readline('nhập số lượng phần tử vec2: '))

vec1 = as.integer(runif(n = n, min=1, max=20))
vec2 = as.integer(runif(n = m, min=1, max=20))
print(vec1)
print(vec2)
```

```r
# Cho biết những phần tử chỉ có trong vec1 mà không có trong vec2
setdiff(vec1, vec2)
```

```r
# Cho biết những phần tử xuất hiện trong cả vec1 và vec2
intersect(vec1, vec2)
```

```r
# Tạo 1 vector mới là vec12 chứa tất cả các phần tử của vec1 và vec2. 
 Phần tử nào xuất hiện nhiều lần nhất ?
vec12 = append(vec1, vec2)
print(vec12)

# Thống kê số lần xuất hiện của từng phần tử.
freq_df = as.data.frame(table(vec12))
cat('\n')
print('Thống kê số lần xuất hiện của từng phần tử: ')
freq_df
for(i in 1:dim(freq_df)[1]){
    print(paste(freq_df[i,1], 'xuất hiện', freq_df[i,2], 'lần'))
}

# Phần tử nào xuất hiện nhiều lần nhất ?
cat('\n')
print('Phần tử nào xuất hiện nhiều lần nhất')
freq_max = freq_df[which.max(freq_df$Freq), ]
freq_max
print(paste('phần tử', freq_max$vec12, 'xuất hiện', freq_max$Freq, 'lần'))
```

```r
# Tạo 1 vector mới là vec_unique chỉ chứa những phần tử duy nhất từ vec12
vec_unique = unique(vec12)
vec_unique
```

#### 7. Hãy lưu dataframe df_golden_delicious vào tập tin golden_delicious.xml

```r
print("Saving file...")
doc <- newXMLDoc()

top <- newXMLNode('golden_delicious', doc=doc)
for (row in 1:nrow(df_golden_delicious)) {
    cdnode <- newXMLNode('fruit', parent=top)
    newXMLNode('fruit_label', df_golden_delicious[row, 'fruit_label'], parent=cdnode)
    newXMLNode('fruit_name', df_golden_delicious[row, 'fruit_name'], parent=cdnode)
    newXMLNode('fruit_subtype', df_golden_delicious[row, 'fruit_subtype'], parent=cdnode)
    newXMLNode('mass', df_golden_delicious[row, 'mass'], parent=cdnode)
    newXMLNode('width', df_golden_delicious[row, 'width'], parent=cdnode)
    newXMLNode('height', df_golden_delicious[row, 'height'], parent=cdnode)
    newXMLNode('color_score', df_golden_delicious[row, 'color_score'], parent=cdnode)
}

cat(saveXML(doc, indent=TRUE,
           prefix="<?xml version=\"1.0\" encoding=\"utf-8\" standalone=\"no\"?>\n"),
   file='golden_delicious.xml')
print("Complete!")
```
