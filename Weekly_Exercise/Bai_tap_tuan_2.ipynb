## Bài tập số 3:


```
create_primes <- function(n){
    primesR <- function(p, i = 1){
        f <- p %% p[i] == 0 & p != p[i]
        if (any(f)){
            p <- primesR(p[!f], i+1)
        }
        p
    }
    primesR(2:n)
}
```

```
primes_list = create_primes(100) # generate vector primes from 2 to n with n < 100
```

```
# Create my_matrix (4x4): each element is a randomly generated prime number.
my_matrix = matrix(sample(primes_list, 16), nrow = 4)
print(my_matrix)

# Sum the elements on the main diagonal of the matrix
print(paste('Sum the elements on the main diagonal of the matrix: ', sum(diag(my_matrix))))


# (Note: Build function to create_matrix, function to test_Prime number, and function to Sum... >> Call fucntion)
# function to create_matrix
create_prime_matrix <- function(max_number, n_rows, n_cols){
    primes_vector = create_primes(max_number)
    return (matrix(sample(primes_vector, n_rows*n_cols), nrow = n_rows, ncol=n_cols))
}

# function to test_Prime number
is_prime <- function(num) {
    if (num == 2) {
        TRUE
    } else if (any(num %% 2:(num-1) == 0)) {
        FALSE
    } else { 
        TRUE
    }
}

# function Sum the elements on the main diagonal of the matrix
sum_diag_matrix <- function(matrixx)
{
    return (sum(diag(matrixx)))
}
```

```
matrix_2 = create_prime_matrix(10000, 8, 8)
matrix_2
```

```
is_prime(3)
is_prime(11)
is_prime(10)
```

```
sum_diag_matrix(matrix_2)
```

```
# Cho dữ liệu mtcars cho sẵn trong R datasets. Đọc dữ liệu. Hiển thị thông tin một số dòng đầu của dữ liệu. In thông tin thống kê chung.
data(mtcars)

head(mtcars)
summary(mtcars)
```

```
# Tạo sub_data chỉ có các thông tin mpg, hp, wt, gear
sub_data = select(mtcars, mpg, hp, wt, gear)
head(sub_data)
```

```
# Tạp cột mới km_per_litre trong sub_data theo công thức: 1 mpg = 0.425 km/l
sub_data$km_per_litre = sub_data$mpg * 0.425
head(sub_data)
```

```
# Tạo sub_data_unique lấy từ mtcars nhưng loại bỏ các dòng trùng (dùng distinct())
print(dim(mtcars))
sub_data_unique = distinct(mtcars)
head(sub_data_unique)
print(dim(sub_data_unique))
```

```
# In sub_data_unique với giá trị được sắp xếp giảm dần theo mpg, tăng dần theo hp
arrange(sub_data_unique, hp, desc(mpg))
```

```
sub_data_unique = cbind(names = rownames(sub_data_unique), sub_data_unique)
head(sub_data_unique)
```

```
# Lọc các xe mà trong tên xe có 'Mazda' trong sub_data_unique. Có bao nhiêu xe thỏa điều kiện này?
filter(sub_data_unique, grepl('Mazda', sub_data_unique$names, fixed=TRUE))
```

```
# Trong sub_data_unique, cho biết giá trị mpg lớn nhất? Có những dòng dữ liệu nào có mpg = mpg lớn nhất, in kết quả.
print(paste('maximum mpg: ', max(sub_data_unique$mpg)))
filter(sub_data_unique, mpg == max(mpg))
```

```
# Lọc các dữ liệu thỏa các điểu kiện sau: mpg >= 25 và cyl=4 và gear=4
filter(sub_data_unique, mpg >= 25, cyl==4, gear==4)
```

```
# Thực hiện việc thống kê theo cyl, từ đó cho biết số lượng từng nhóm, cyl, hp lớn nhất, mpg nhỏ nhất
summary(sub_data_unique$cyl)

df_group = group_by(sub_data_unique, cyl)
summarize(  df_group, 
            count=n(),
            max.hp = max(sub_data_unique$hp, na.rm=T),
            min.mpg = min(sub_data_unique$mpg, na.rm=T)
)
```

```
# Trong sub_data, cho biết xe nào có km_per_litre lớn nhất, nhỏ nhất?
filter(sub_data, km_per_litre == max(km_per_litre))
filter(sub_data, km_per_litre == min(km_per_litre))
```
