```
# Tìm xem từ vừa nhập có là phần tử nào trong danh sách phần tử trên hay không? Nếu tìm thấy thì thông báo là "Found!!!" 
# ngược lại thì thông báo là "Not found!!!"
if (word %in% (unlist(split_str_list))){
    print('Found!!!')
} else{
    print('Not found!!!')
}

```

## Bài tập số 3:
### Viết chương trình cho phép người dùng:
- Nhập vào một số n với điều kiện sau: 10<=n<=20. Nếu số người dùng nhập không thỏa điều kiện thì thông báo cho người dùng biết và yêu cầu nhập lại. Nếu thỏa điều kiện thì qua 2.
- Tạo một dãy số v1 có n số mà mỗi số là một số ngẫu nhiên từ 1 đến 10. Ví dụ: v1: 5, 5, 5, 9, 9, 8, 5, 9, 9, 10, 7, 7
- Viết một function tạo dãy số mới v2 từ dãy số v1 nói trên theo điều kiện sau: phần tử đầu tiên và cuối cùng của v2 bằng v1, các phần tử khác thì v2[i] = v1[i] + v1[i-1] + v1[i+1] nếu cả v1[i-1] và v1[i+1] cùng chẵn hoặc cùng lẻ, ngược lại v2[i] = v1[i]. Ví dụ: v2: 5, 15, 19, 23, 9, 22, 5, 23, 9, 26, 7, 7
- Gọi function vừa viết và in kết quả v2.
- Viết function Cho biết trong một dãy số có số chính phương nào hay không?
- Gọi function vừa viết với input là v2 và in kết quả.

```
# Nhập vào một số n với điều kiện sau: 10<=n<=20. Nếu số người dùng nhập không thỏa điều kiện thì thông báo cho người dùng biết và yêu cầu nhập lại. 
# Nếu thỏa điều kiện thì qua 2.
while(TRUE){
    n = as.integer(readline('Entrez n (10<=n<=20): '))
    if(n>=10 && n<=20){break}
}
```

```
# Tạo một dãy số v1 có n số mà mỗi số là một số ngẫu nhiên từ 1 đến 10. Ví dụ: v1: 5, 5, 5, 9, 9, 8, 5, 9, 9, 10, 7, 7
v1 = as.integer(runif(n = n, min=1, max=10))
print(v1)
```

```
# Viết một function tạo dãy số mới v2 từ dãy số v1 nói trên theo điều kiện sau: 
# phần tử đầu tiên và cuối cùng của v2 bằng v1, 
# các phần tử khác thì v2[i] = v1[i] + v1[i-1] + v1[i+1] nếu cả v1[i-1] và v1[i+1] cùng chẵn hoặc cùng lẻ, 
# ngược lại v2[i] = v1[i]. Ví dụ: v2: 5, 15, 19, 23, 9, 22, 5, 23, 9, 26, 7, 7

create_vector = function(vector1){
    vector2 = vector1
    for(i in 2:(length(vector1)-1)){
        if((vector1[i-1] %% 2 == 0 && vector1[i+1] %% 2 == 0) || (vector1[i-1] %% 2 != 0 && vector1[i+1] %% 2 != 0)){
            vector2[i] = vector1[i] + vector1[i-1] + vector1[i+1]
        } else {
            vector2[i] = vector1[i]
        } 
    }
    return (vector2)
}
```

```
# Gọi function vừa viết và in kết quả v2
v2 = create_vector(v1)
print(v2)
```

```
# Viết function Cho biết trong một dãy số có số chính phương nào hay không?
isPerfectSquare = function(number){
    if (number >=0){
        sr = as.integer(sqrt(number))
        return ((sr*sr)==number)
    }
    return (FALSE)
}

isPerfectSquare_Vector = function(vec){
    vec_check = isPerfectSquare(vec[1])
    for(i in 2:length(vec)){
        vec_check = append(vec_check, isPerfectSquare(vec[i]))
    }
    return (vec_check)
}
```
