### Exercise 2: Đọc tập tin, xử lý, ghi nội dung vào tập tin

```r
df <- read.table("data/Table0.txt")
print("Content of file:")
print(df)
```

```r
names(df) <- c('Name', 'Age', 'Height', 'Weight', 'Sex')
print(df)
```

```r
#dat rowname la du lieu cot Name
row.names(df)<-df$Name
print(df)
```

```r
#sau do bo di cot name thua
df$Name <- NULL
print("Content of data frame:")
print(df)
```

```r
df_new <- read.table("data/Table0_new.txt")
print("Read saved file:")
print(df_new)
```

### Exercise 3: Đọc, ghi tập tin txt

```r
df <- read.table("data/Table6.txt",
                skip = 1,
                header = TRUE, 
                flush = TRUE,
                comment.char = "@",
                stringsAsFactors = FALSE)
                
print("Content of file after reading:")
print(head(df))
```

```r
print("Before dropping: ")
print(paste("Rows:", nrow(df)))
```

```r
library(tidyverse)
```

```r
df = unique(df)
print("After dropping: ")
print(paste("Rows:", nrow(df)))
```

### Exercise 5: Đọc tập tin csv, xử lý, lưu vào tập tin xlsx

```r
library("xlsx")
```

```r
data <- read.csv("data/medals.csv")
head(data)
```

```r
print(paste("Data type: ",class(data)))
print(paste("Number of rows:",ncol(data)))
print(paste("Number of cols:",nrow(data)))
print(paste("Name of columns", toString(names(data))))
```

```r
data_1932 = subset(data, Year == 1932 & Sport=='Skiing')
print("Data set of year = 1932 and Sport = Skiing:")
data_1932
```

```r
# ghi bo du lieu nay vao file excel medals_new.xlsx
print("Write to file...")
write.xlsx(data_1932, file = "data/medals_new.xlsx", sheetName = "1932_skiing_2", append = TRUE)
print("Finish!")
```
