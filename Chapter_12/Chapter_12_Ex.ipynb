### **Exercise 1: Chol - ggplot2**

```r
library(ggplot2)
library(ggExtra)
library(tidyverse)
```

```r
# Load in `chol` data
chol <- read.table("data/chol.txt", header = TRUE)
# Inspect first rows of `chol` with `head()`
head(chol)

# Summary with `summary()`
summary(chol)

# Structure of `chol` with `str()`
str(chol)
```

```r
ggplot(data=chol, aes(chol$AGE)) + geom_histogram(bins=30)
```

```r
ggplot(data=chol, aes(x=chol$AGE)) +
        geom_histogram(breaks=seq(20, 50, by=2), 
        col="red", 
        fill="green", 
        alpha = .2) +
        labs(title="Histogram for Age", x="Age", y="Count") +
        xlim(c(18,52)) +
        ylim(c(0,30)
)
```

```r
ggplot(data=chol, aes(chol$AGE)) +
        geom_histogram(breaks=seq(20, 50, by=2), 
        col="red", 
        fill="green"
)
```

```r
ggplot(data=chol, aes(chol$AGE)) +
        geom_histogram(breaks=seq(20, 50, by=2), 
        col="red", 
        aes(fill=..count..)) +
        scale_fill_gradient("Count", low="green", high="red"
)
```

```r
ggplot(data=chol, aes(chol$AGE)) +
        geom_histogram(aes(y =..density..), 
        breaks=seq(20, 50, by = 2), 
        col="red", 
        fill="green", 
        alpha=.2
) + geom_density(col=2) +
    labs(title="Histogram for Age", x="Age", y="Count")
```

```r
options(scipen=999) # turn-off scientific notation like 1e+48
library(ggplot2)
theme_set(theme_bw()) # pre-set the bw theme.

# Scatterplot
gg <- ggplot(data = chol, aes(x=WEIGHT, y=HEIGHT)) +
            geom_point() +
            geom_smooth(method="lm", se=F) + 
            labs(y="HEIGHT", 
            x="WEIGHT", 
            title="WEIGHT Vs HEIGHT", 
            caption = "Source:CHOL ")
plot(gg)
```

```r
theme_set(theme_bw()) # pre-set the bw theme.

# Scatterplot
gg <- ggplot(data = chol, aes(x=WEIGHT, y=HEIGHT)) +
            geom_point() +
            geom_smooth(method="lm", se=F) + 
            labs(y="HEIGHT", 
            x="WEIGHT", 
            title="WEIGHT Vs HEIGHT", 
            caption = "Source:CHOL ")
            
ggMarginal(gg, type = "histogram", fill="transparent", bins = 30)
ggMarginal(gg, type = "boxplot", fill="transparent")
```

```r
pie <- ggplot(chol, aes(x = "", fill = factor(BLOOD))) +
            geom_bar(width = 1) +
            theme(axis.line = element_blank(), 
            plot.title = element_text(hjust=0.5)) +
            labs(fill="class", 
            x=NULL, 
            y=NULL, 
            title="Pie Chart of BLOOD", 
            caption="Source: CHOL")
 
pie + coord_polar(theta = "y", start=0)
```

### **Exercise 2: Housing Prices**

```r
housing = read.csv("data/landdata_states.csv")
head(housing)
summary(housing)
str(housing)
```

```r
ggplot(housing, aes(x = Home.Value)) + geom_histogram(bins = 30)
```

```r
hp2001Q1 <- filter(housing, Date == 2001.25) 
ggplot(hp2001Q1,
        aes(y = Structure.Cost, x = Land.Value)) +
        geom_point()
```

```r
p1 = ggplot(hp2001Q1,
            aes(y = Structure.Cost, x = log(Land.Value))) +
            geom_point(color="red")
p1
```

```r
p1 + geom_text(aes(label=State), size = 3)
```

```r
p1 + geom_point(aes(color=Home.Value, shape = region))
```
