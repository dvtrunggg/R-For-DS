### Câu 1

```r
# tỷ lệ bao nhiêu phần trăm người có IQ nhỏ hơn 125
p_lower_125 = pnorm(125, mean=100, sd=15)
print(paste('tỷ lệ bao nhiêu phần trăm người có IQ nhỏ hơn 125', p_lower_125))
```

```r
# tỷ lệ bao nhiêu phần trăm người có IQ lớn hơn 110?
p_higher_110 = pnorm(110, mean=100, sd=15, lower.tail=FALSE)
print(paste('tỷ lệ bao nhiêu phần trăm người có IQ lớn hơn 110', p_higher_110))
```

```r
#tỷ lệ bao nhiều phần trăm người có IQ trong khoảng từ 110 và 125 ?
p_lower_110 = pnorm(110, mean=100, sd=15)
p_between_110_125 = p_lower_125 - p_lower_110
print(paste('phần trăm người có IQ trong khoảng từ 110 và 125', p_between_110_125))
```
