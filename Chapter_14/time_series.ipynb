## **HoltWinter**

```r
forcast.rainfall = HoltWinters(ts.rainfall, beta = F, gamma = F)
```

```r
forcast.rainfall
```

```r
# alpha càng tiến về 1 thì càng có ý nghĩa để dự đoán giá trị tiếp, tiến về 0 thì ngược lại
```

```r
forcast.rainfall$SSE
```

```r
rmse = forcast.rainfall$SSE/length(ts.rainfall)^0.5
print(paste('RMSE', rmse))
```

### **Dự đoán**

```r
library('forecast')
```

```r
forcast.rainfall.new = forecast:::forecast.HoltWinters(forcast.rainfall, h = 5)
```

```r
forcast.rainfall.new
```

```r
# => khoảng cách lớn wa :)))
```

```r
plot(forcast.rainfall.new)
```

```r
# phần dự đoán (màu xanh) còn khoảng tin cậy 80-95% là vùng màu xám
# => từ biểu đồ: nhìn chung là sai số cao:))
# => SES ko phải là phương pháp thích hợp cho dữ liệu nàyy
```

### **HES**

```r
HES.forecast = HoltWinters(log(ts.rainfall), gamma = F)
HES.forecast
```

```r
HES.forecast$SSE
```

```r
rmse = HES.forecast$SSE/length(ts.rainfall)^0.5
print(paste('RMSE', rmse))
```

```r
plot(HES.forecast)
```

```r
HES.forecast.new = forecast:::forecast.HoltWinters(HES.forecast, h=5)
HES.forecast.new
```

```r
plot(HES.forecast.new)
```
