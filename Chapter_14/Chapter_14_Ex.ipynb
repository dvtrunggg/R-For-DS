### **Exercise 1: Kings**
- Cung cấp tập tin kings.dat.txt
- Đọc dữ liệu từ tập tin và in dữ liệu
- Chuyển dữ liệu này thành Time Series object => in Time Series object
- Vẽ Time Series object vừa tạo


```r
# https://drive.google.com/file/d/1GTWrRyXgo7OPj2Fz3qHOtLuorYgn6dup/view?usp=sharing
system('gdown --id 1GTWrRyXgo7OPj2Fz3qHOtLuorYgn6dup')
```

```r
kings <- scan("kings.dat.txt",skip=3)
print(kings)
```

```r
# Convert it to a time series object.
kings.timeseries <- ts(kings,start = 1, end = length(kings), frequency = 1)
print(kings.timeseries)
```

```r
# Plot a graph of the time series.
plot(kings.timeseries, main='Age of Death of Successive Kings of England', 
    col='blue',
    xlab='Kings', ylab='Age')
```

- Mô hình trên không có chu kỳ rõ ràng
- Độ dao động ngẫu nhiên có vẻ độc lập với thời gian

- Cung cấp tập tin mybirths.dat.txt
- Đọc dữ liệu từ tập tin, in dữ liệu
- Chuyển dữ liệu này thành Time Series object => in Time Series object
- Vẽ Time Series object vừa tạo
- Thực hiện việc dự báo và vẽ biểu đồ so sánh với thực tiễn

```r
# https://drive.google.com/file/d/18G7zQ_sbSVj2AdfHmN44RKM4IcKJrN89/view?usp=sharing
system('gdown --id 18G7zQ_sbSVj2AdfHmN44RKM4IcKJrN89')
```

```r
births <- scan("mybirths.dat.txt",skip=3)
print(births)
```

```r
# Convert it to a time series object.
print("Bitrhs Time Series:")
births.timeseries <- ts(births, start = c(1946,1), frequency = 12)

# Print the timeseries data.
births.timeseries
```

```r
plot(births.timeseries, 
    main='Births per month in New York city, from January 1946 to December 1959',
    col='red',
    xlab='Year', ylab='Births'
)
```

<font color = 'red'>**Nhận xét:**</font>
- Có thể thấy từ biểu đồ này một chu kỳ mang tính mùa và xu hướng năm
- Ta nhận thấy, số em bé sinh ra tăng dần theo các năm từ 1950 đổ về sau (theo xu hướng bùng nổ dân số)
- Tuy nhiên ở thời điểm từ nam9 1946 đến 1950, số lượng em bé sinh ra giảm do suy thoái kinh tế Mỹ giai đoạn 1947 vì đây là thời điểm hậu chiến tranh thế giới lần II - giai đoạn suy sụp kinh tế tồi tệ nhất trong lịch sử hiện đại

### **Forecasting**

```r
# Forecasting
births.timeseries.forecasts <- HoltWinters(births.timeseries)
print(births.timeseries.forecasts)
```

<font color = 'red'>**Nhận xét:**</font>
- alpha ~ 0.7: khá cao, cho thấy giá trị hiện tại phụ thuộc vào các giá trị gần đây và các giá trị
trong quá khứ gần
- beta ~ 0: độ dốc của trend không cập nhập trong thời gian qua
- gamma ~ 0.7: ảnh hưởng của yếu tố mùa tại thời điểm hiện tại có phụ thuộc vào các giá trị gần

```r
options(repr.plot.width = 15, repr.plot.height = 8)
plot(births.timeseries.forecasts, 
    main = "Births: observer vs forecast",
    xlab='Year', ylab='Births'
)
```

<font color = 'red'>**Nhận xét:**</font>
- Nhìn tổng quan thì các giá trị forecast khá khớp và theo xu hướng của các giá trị quan trắc (observer)
