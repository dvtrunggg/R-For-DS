## **Exercise 1: Tạo dataframe**

```r
Name <- c("Alex", "Lilly", "Mark", "Oliver", "Martha", "Lucas", "Caroline")
Age <- c(25, 31, 23, 52, 76, 49, 26)
Height <- c(177, 163, 190, 179, 163, 183, 164)
Weight <- c(57, 69, 83, 75, 70, 83, 53)
Sex <- as.factor(c("F", "F", "M", "M", "F", "M", "F"))
df <- data.frame (row.names = Name, Age, Height, Weight, Sex)

print(df)
```

```r
print(paste("Levels of Sex:",toString(levels(df$Sex))))
```

```r
Working <- c("Yes","No","No","Yes","Yes","No","Yes")
df2 <- data.frame(row.names = Name,Working) #Name has been already defined in df
print(df2)
```

```r
# tao data frame moi chua thong tin cua 2 data frame tren
df <- cbind(df,df2)
print("Combining 2 data frame:")
print(df)
```

```r
#co bao nhieu dong va bao nhieu cot trong data frame nay?
print(paste("Dong:", dim(df)[1], ", Cot:", dim(df)[2]))
# cho biet kieu du lieu cua tung cot
print(str(df))
```

## **Exercise 2: Tạo và làm việc với dataframe**

## **Exercise 4: Làm việc với dataframe**

```r
# tao data frame tu state.abb, state.area, state.division, state.name, state.regi
# row names la ten cua states.
df <- data.frame(state.abb, state.area, state.division, state.region, 
                row.names = state.name)
print("Data frame:")
print(head(df))
```

```r
# doi ten tat cac cac cot chi chua 3 ky tu sau dau .
colnames(df) <- substr(colnames(df), 7, 9)
# sau khi doi ten
print("After rename:")
print(head(df))
```

```r
# tao data frame chua state.x77 va du lieu moi tao
df_new <- cbind(state.x77, df)
print("New data frame:")
print(head(df_new))
```

```r
# loai bo cot div
df_new$div <- NULL
# tiep tuc loai bo cac cot Life Exp, HS Grad, Frost, abb, va are
df_new <- subset(df_new,select = -c(4, 6, 7, 9, 10))
print("After drop div, Life Exp, HS Grad, Frost, abb, and are:")
print(head(df_new))
```

```r
# them cot phan loai mu chu Illiteracy.Levels voi mo ta sau: <1 Low, <2 Some
# con lai la High
df_new$Illiteracy.Levels <- ifelse(df_new$Illiteracy >= 0 &
                                    df_new$Illiteracy < 1, 
                                    "Low",
                                    ifelse(df_new$Illiteracy >= 1
                                    & df_new$Illiteracy < 2, 
                                    "Some",
                                    "High")
)

print("After insert Illiteracy.Levels:")
print(head(df_new))
```

```r
# cho biet co bao nhieu vung va la nhung vung nao?
print(paste("Number of Regions:", nlevels(df_new$reg), ". There are:", toString(levels(df_new$reg))))
```

```r
# Cho dữ liệu Obesity_data.csv. Đọc dữ liệu vào dataframe data. In head của data
data <- read.csv('Obesity_data.csv')
head(data)
```

```r
#  Tạo dataframe data1 chỉ chứa các cột id, gender, height, weight, bmi. In head của data1
data1 = select(data, gender, height, weight, bmi, age)
print(head(data1))
```

```r
# Với data1, lọc ra những dòng có bmi >=18.5 và <=24.9 và đưa vào dataframe data2
data2 = filter(data1, bmi >=18.5, bmi <=24.9)
print(head(data2))
print(paste("Rows have bmi >=18.5 and <=24.9: ", nrow(data2)))
```

```r
# Với data1, tạo ta biến mới height_m = height/100
data1 = mutate(data1, height_m = height/100)
head(data1)
```

```r
# Với data1, sắp tăng dần theo bmi
head(arrange(data1, bmi))
```

```r
# Với data1, Tính giá trị trung bình của height, weight theo gender. Tạo ra các biến mới chứa
# giá trị trung bình là mean.height, mean.weight

group = group_by(data1, gender)
df_summary = summarize(group,
                        count = n(),
                        mean.height = mean(height, na.rm = T),
                        mean.weight = mean(weight, na.rm = T) 
 )
 df_summary
```

```r
# Với data1, đếm số lượng theo gender và age
count(data1, gender)
```

```r
# viết cho vui ạ ^^
group_gender_age = count(data1, gender, age)

print(head(group_gender_age))
print(tail(group_gender_age))
```

```r
# Chọn 10 mẫu ngẫu nhiên từ data1 và đưa vào data3
data3 = data1 %>% sample_n(10)
data3
```
