### **Exercise 2: PIE 3D**

```
Tree_specices <- c("Beech", "Chestnut", "Hawthorn",
                "Red maple", "Witch Hazel", "Yellow Birch")
number_of_trees <- c(2, 2, 1, 4, 1, 1)
piepercent<- round(100*number_of_trees/sum(number_of_trees), 1)
```

```
pie3D(number_of_trees, labels = piepercent, col = rainbow(length(number_of_trees)), main = "Number of Trees of each Species") 
legend("topright", Tree_specices, cex =0.7, fill = rainbow(length(number_of_trees)))
dev.off()
```

```
barplot(number_of_trees, names.arg = Tree_specices,
        xlab = "Species", ylab = "Numbers", col = "#43b062",
        main = "Number of Trees of each Species")
```

### **Exercise 7: Leaves had fallen – Line graph**

```
years <- c(2014, 2015, 2016, 2017)
first_fallen <- c(309, 271, 263, 255)
one_hundred_fallen <- c(309,314,312,304)
```

### **Exercise 8: Tech_com – line graph**

```
# https://drive.google.com/file/d/1SsTefYupNqe05Mqm08VIABmxSaeU6W8n/view?usp=sharing
system('gdown --id 1SsTefYupNqe05Mqm08VIABmxSaeU6W8n')
```

```
data_tech_com <- read.csv(file= "Tech_comp.csv")
print(data_tech_com)
```

```
Men..non..tech <- data_tech_com$Men..all. - data_tech_com$Men..tech.
Women..non..tech <- data_tech_com$Women..all - data_tech_com$Women..tech.
```

```
g_range <- range(0,data_tech_com$Men..tech., data_tech_com$Women..tech.,
                Men..non..tech, Women..non..tech)
```

```
# Plot the line chart.
plot(data_tech_com$Men..tech., type="o", col="blue", ylim = g_range, 
    xlab = "Scocial Network", ylab = "Numbers",
    main = "Tech vs Nontech in Company", xaxt='n'
)
axis(side=1, at=seq(1,4,1), labels = data_tech_com$Company)

# Create box around plot
box()

# lines
lines(Men..non..tech, type = "o", col = "red")
lines(data_tech_com$Women..tech., type = "o", col = "green")
lines(Women..non..tech, type = "o", col = "black")

# insert legend for chart
legend("bottomright",
        legend = c("Men tech", "Men non tech", "Women tech", "Women", 
                    cex = 0.8, 
                    fill = c("blue","red", "green","black")
                  )
)
```
